## Resolução Numérica:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#funcoes
def f(x, y):
  return -1.2*y + 7*np.exp(-0.3*x)

def y_exato(x):
  return 70/9*np.exp(-0.3*x) - 43/9*np.exp(-1.2*x)

def euler_modificado(f, x0, y0, h, n):
  x = np.zeros(n+1)
  y = np.zeros(n+1)
  x[0] = x0
  y[0] = y0
  for i in range(n):
    y_pred = y[i] + h*f(x[i], y[i])
    x[i+1] = x[i] + h
    y[i+1] = y[i] + h/2*(f(x[i], y[i]) + f(x[i+1], y_pred))
  return x, y

def euler_explicito(f, x0, y0, h, n):
  x = np.zeros(n+1)
  y = np.zeros(n+1)
  x[0] = x0
  y[0] = y0
  for i in range(n):
    y[i+1] = y[i] + h*f(x[i], y[i])
    x[i+1] = x[i] + h
  return x, y
```

```python
#paramentros
x0 = 0
y0 = 3
h = 0.5
n = 5
```

```python
#Solução numérica (Euler Modificado)
x_mod, y_mod = euler_modificado(f, x0, y0, h, n)
```

```python
#Solução numérica (Euler Explícito)
x_exp, y_exp = euler_explicito(f, x0, y0, h, n)
```

```python
#Solução exata
x_exato = np.linspace(x0, x0 + n*h, 6)
y_exato = y_exato(x_exato)
```

```python
plt.plot(x_mod, y_mod, 'o-', label='Euler Modificado')
plt.plot(x_exp, y_exp, 's-', label='Euler Explícito')
plt.plot(x_exato, y_exato, label='Solução Exata')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparação entre Métodos de Euler e Solução Exata')
plt.legend()
plt.grid(True)
plt.show()
```
