Resoluçao manual para as duas primeiras iteraçoes

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Criando a funcao e sua derivada
def f(x):
  return 8 - 4.5*(x - math.sin(x))

def df(x):
  return (-4.5) + 4.5*(math.cos(x))
```

Definimos uma funcao para a equaçao do problema e sua derivada, pois o metodo de Newton necessita delas para calcular o proximo valor

```python
# Metodo de Newton

def newton(error, it, x0):
    i = 1
    while i <= it:
        x1 = x0 - f(x0)/df(x0)     # calcula o proximo x
        if abs(x1 - x0) < error:   # verifica o erro
            return x1, i
        i += 1
        x0 = x1
    return x1, i
```

Dentro desta função, temos um loop que executa até que o número máximo de iterações seja atingido ou a diferença entre duas aproximações consecutivas seja menor que o erro especificado.

```python
# Condiçoes iniciais

erro_lim = 0.0001 #erro relativo maximo
i = 10   #maximo de iteraçoes
x0 = 2   # valor inicial
```

```python
# Chamada da funcao

resposta, iteracoes = newton(erro_lim, i, x0)

print(f"A soluçao da equaçao é {resposta:.6f} e foi encontrada após {iteracoes} iterações")
```

Aqui, chamamos a função newton passando os valores inicias dados no problema para encontrar a raiz e o número de iterações necessarias para acha-la.
