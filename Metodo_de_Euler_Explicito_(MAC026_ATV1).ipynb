```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando a função
def f(x, y):
    return -1.2 * y + 7 * np.exp(-0.3 * x)

def MetodoEuler(h, x_final):
    # Condições iniciais
    x0, y0 = 0, 3
    x = [x0]
    y = [y0]

    while x[-1] < x_final:
        y1 = y[-1] + h * f(x[-1], y[-1])  #Calculando os valores de Y_i+1
        x.append(x[-1] + h)
        y.append(y1)

    return x, y
```

```python
# Passo h = 0,5
h = 0.5
x1, y1 = MetodoEuler(h, 2.5)
print("Valores com passo 0.5:")
for x, y in zip(x1, y1):
    print(f"x = {x}, y = {y:.3f}")
```

```python
# Passo h = 0,1
h = 0.1
x2, y2 = MetodoEuler(h, 2.5) # novos vetores para armazenar os dados de h=0.1
print("Valores com passo 0.1:")
for x, y in zip(x2, y2):
    print(f"x = {x:.1f}, y = {y:.3f}")
```

```python
# Passo h = 0,01
h = 0.01
x3, y3 = MetodoEuler(h, 2.49)
print("Valores com passo 0.01:")
for x, y in zip(x3, y3):
    print(f"x = {x:.2f}, y = {y:.3f}")
```

```python
# Calculado os valores com a solucao exata
y_exato = []

def solucaoExata(x):
  return ((70/90) * np.exp(-0.3 * x)) - ((43/9) * np.exp(-1.2 * x))

for i in range (len(x3)):
  y_exato.append(solucaoExata(x3[i]))

for x, y in zip(x3, y_exato):
    print(f"x = {x:.2f}, y = {y:.3f}")
```

```python
# plotando o grafico

plt.figure()
plt.plot(x1, y1, label='Passo = 0.5')
plt.plot(x2, y2, label='Passo = 0.1')
plt.plot(x3, y3, label='Passo = 0.01')
plt.plot(x3, y_exato, label='Soluçao Exata')

plt.title('Solução numerica X Solução exata')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
```
